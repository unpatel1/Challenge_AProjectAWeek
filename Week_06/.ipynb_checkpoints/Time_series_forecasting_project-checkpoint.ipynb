## --- Treating missing values ---

```python
# Ckecking the missing values
plt.rcParams['figure.figsize'] = (15, 5)
plt.plot(data, color = 'black')
plt.title("Number of Customers Visiting an Ice Cream Shope Since 1949")
plt.show()
```

### Imputing missing values in a time series
* 1. Mean imputation  
* 2. Last observation carried forward: Impute missing values with the previous value in the data  
* 3. Linear interpolation: Draw a straight line joining the previous and next points of the missing values  
* 4. Seasonal interpolation: Impute the missing values with the average of corresponding data points from the previous seasonal data and the next seasonal data

#### 1. Mean imputation

```python
plt.rcParams['figure.figsize'] = (15, 5)
data['Customers_mean'] = data['Customers'].fillna(data['Customers'].mean())
plt.plot(data['Customers_mean'], color = 'blue')
plt.title("Missing Values - Mean Imputation")
plt.show()
```

#### 2. Last observation carried forward (LOCF)

```python
plt.rcParams['figure.figsize'] = (15, 5)
data['Customers_locf'] = data['Customers'].bfill()
plt.plot(data['Customers_locf'], color = 'black')
plt.title("Missing Values - Last Observation Carried Forward (LOCF)")
plt.show()
```

#### 3. Linear interpolation

```python
plt.rcParams['figure.figsize'] = (15, 5)
data['Customers_linear'] = data['Customers'].interpolate(method = 'linear')
plt.plot(data['Customers_linear'], color = 'green')
plt.title("Missing Values - Linear Interpolation")
plt.show()
```

## --- Time series decomposition ---

```python
import statsmodels.tsa.api as tsa
import statsmodels as sm
```

### Additive seasonal decomposition

```python
plt.rcParams['figure.figsize'] = (15, 12)

decomposition = sm.tsa.seasonal.seasonal_decompose(data['Customers_linear'], model = 'additive')
decomposition.plot()

plt.show()
```

**Comments:**
* There is some pattern in the residual component.
* We can check this with multiplicative decomposition method.

### Multiplicative seasonal decomposition

```python
plt.rcParams['figure.figsize'] = (15, 12)

decomposition = sm.tsa.seasonal.seasonal_decompose(data['Customers_linear'], model = 'multiplicative')
decomposition.plot()

plt.show()
```

**Comments:**
* There is no clear pattern visible in thr residual component.
* We will consider the Multiplicative seasonal decomposition as it gives a better result.
* We will need a time series model that would capture all the three components of a time series data, that is level, trend and seasonality.
* The models that capture all three time series components are 'Holt Winter Exponential Smoothing' and 'SARIMA' model.

## --- Model building ---

### Splitting data into train and test sets

```python
length_train = 115
train = data.iloc[:length_train, :]
test = data.iloc[length_train:, :]
```

### Holt Winter Exponential Smoothing

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

model = ExponentialSmoothing(train['Customers'], seasonal_periods = 12, trend = 'multiplicative', seasonal = 'additive')
model_fit = model.fit(smoothing_level = 0.2, smoothing_slope = 0.04)
model_fit.params

y_holtwinter = test.copy()
y_holtwinter['Holtwinter_forecast'] = model_fit.forecast(36)
```

```python
plt.rcParams['figure.figsize'] = (15, 5)

plt.plot(train['Customers'], label = 'Train')
plt.plot(test['Customers'], label = 'Test')
plt.plot(y_holtwinter['Holtwinter_forecast'], label = 'Holts winters exponential forecast')

plt.legend()
plt.title('Holts Winters Exponential Method')
plt.show()
```

### Evaluating the model

```python
from sklearn.metrics import mean_squared_error

mfe = np.round(np.mean(test['Customers'] - y_holtwinter['Holtwinter_forecast']),
               2) # Mean forecast error
mae = np.round(np.mean(np.abs(test['Customers'] - y_holtwinter['Holtwinter_forecast'])),
               2) # Mean absolutely error
mape = np.round(np.mean(np.abs(test['Customers'] - y_holtwinter['Holtwinter_forecast']
                              )/test['Customers'])*100,2) # Mean absolutely percentage error
rmse = np.round(np.sqrt(mean_squared_error(test['Customers'], y_holtwinter['Holtwinter_forecast'])),
                2) # Root mean square error
```

```python
result = pd.DataFrame({'mfe':mfe, 'mae':mae, 'mape':mape, 'rmse':rmse}, index = [0])
result
```
