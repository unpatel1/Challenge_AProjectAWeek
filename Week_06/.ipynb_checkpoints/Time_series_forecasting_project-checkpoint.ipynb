### Imputing missing values in a time series
* 1. Mean imputation  
* 2. Last observation carried forward: Impute missing values with the previous value in the data  
* 3. Linear interpolation: Draw a straight line joining the previous and next points of the missing values  
* 4. Seasonal interpolation: Impute the missing values with the average of corresponding data points from the previous seasonal data and the next seasonal data

#### 1. Mean imputation

```python
plt.rcParams['figure.figsize'] = (15, 5)
data['Customers_mean'] = data['Customers'].fillna(data['Customers'].mean())
plt.plot(data['Customers_mean'], color = 'blue')
plt.title("Missing Values - Mean Imputation")
plt.show()
```

#### 2. Last observation carried forward (LOCF)

```python
plt.rcParams['figure.figsize'] = (15, 5)
data['Customers_locf'] = data['Customers'].bfill()
plt.plot(data['Customers_locf'], color = 'black')
plt.title("Missing Values - Last Observation Carried Forward (LOCF)")
plt.show()
```

#### 3. Linear interpolation

```python
plt.rcParams['figure.figsize'] = (15, 5)
data['Customers_linear'] = data['Customers'].interpolate(method = 'linear')
plt.plot(data['Customers_linear'], color = 'green')
plt.title("Missing Values - Linear Interpolation")
plt.show()
```

## --- Model building ---

### Splitting data into train and test sets

```python
length_train = 115
train = data.iloc[:length_train, :]
test = data.iloc[length_train:, :]
```

### Holt Winter Exponential Smoothing

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

model = ExponentialSmoothing(train['Customers'], seasonal_periods = 12, trend = 'multiplicative', seasonal = 'additive')
model_fit = model.fit(smoothing_level = 0.2, smoothing_slope = 0.04)
model_fit.params

y_holtwinter = test.copy()
y_holtwinter['Holtwinter_forecast'] = model_fit.forecast(36)
```

```python
plt.rcParams['figure.figsize'] = (15, 5)

plt.plot(train['Customers'], label = 'Train')
plt.plot(test['Customers'], label = 'Test')
plt.plot(y_holtwinter['Holtwinter_forecast'], label = 'Holts winters exponential forecast')

plt.legend()
plt.title('Holts Winters Exponential Method')
plt.show()
```

### Evaluating the model

```python
from sklearn.metrics import mean_squared_error

# Mean forecast error
mfe_Holts_Winters = np.round(np.mean(test['Customers'] - y_holtwinter['Holtwinter_forecast']), 2) 

# Mean absolutely error
mae_Holts_Winters = np.round(np.mean(np.abs(test['Customers'] - y_holtwinter['Holtwinter_forecast'])), 2)

# Mean absolutely percentage error
mape_Holts_Winters = np.round(np.mean(np.abs(test['Customers'] - y_holtwinter['Holtwinter_forecast'])/test['Customers'])*100, 2) 

# Root mean square error
rmse_Holts_Winters = np.round(np.sqrt(mean_squared_error(test['Customers'], y_holtwinter['Holtwinter_forecast'])), 2) 
```

```python
result_Holts_Winters = pd.DataFrame({'mfe':mfe_Holts_Winters, 'mae':mae_Holts_Winters, 'mape':mape_Holts_Winters, 'rmse':rmse_Holts_Winters}, index = [0])
result_Holts_Winters
```

### SARIMA model
#### Data processing for the SARIMA model
SARIMA model involves all the three components of the time series data which are level, trend and seasonality.

**Assumptions for building SARIMA model**  

Assumption 1. Stationarity: Statistical properties of a process generating a time series do not change over time  
a. Mean  
b. Variance  
c. covariance  
Assumption 2. Autocorrelation

#### Assumption 1. Checking stationarity of a time series  
* We can use the following two statistical tests to check the stationarity of a time series
1. ADF test
2. KPSS test

#### ADF test
* **p-value > 0.05:** Fail to reject the null hypothesis (H0), the data has a unit root and is `non-stationary`
* **p-value <= 0.05:** Reject the null hypothesis (H0), the data does not have a unit root and is `stationary`

```python
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(data['Customers_linear'], autolag = 'AIC')
print(f'* ADF Statistics: {result[0]}')
print(f'* p-value: {result[1]:.20f}')
print(f'* # of lags used: {result[2]:.20f}')
# print(f'* # of observations used for the ADF regression and calculation of the critical values: {result[3]}')
print('* critical values - Critical values for the test statistic at the 1%, 5%, and 10% levels. Based on MacKinnon: \n')
for key, value in result[4].items():
    print('  Critical value:', end = "")
    print(f' {key}, {value}')
```

P-value is 0.99 which is > 0.05 so we fail to reject the null hypothesis (H0), the data does not have a unit root and so the series is non-stationary.

#### KPSS test
* **Null hypothesis (H0):** The series is `stationary` when p-value > 0.05  
* **Alternate hypothesis (H1):** The series is `non-stationary` when p-value <= 0.05

```python
from statsmodels.tsa.stattools import kpss
```

```python
result = kpss(data['Customers_linear'])
print(f'* KPSS statistics: {result[0]}')
print(f'* P-value: {result[1]}')
print(f'* # of lags: {result[2]}')
print('* The critical values at 10%, 5%, 2.5% and 1%. Based on Kwiatkowski et al. (1992).: \n')
for key, value in result[3].items():
    print('  Critical value:', end = "")
    print(f' {key}, {value}')
```

P-value is 0.01 which is < 0.05 so we reject the null hypothesis (H0), the data does not have a unit root and so the series is non-stationary.

### Converting non-stationary into stationary
There are two tools to convert a non-stationary series into a stationary series.
1. Box Cox transformation
2. Differencing

#### Box Cox transformation

```python
# Importing boxcox from scipy
from scipy.stats import boxcox
```

```python
# Applying the boxcox transformation with lambda as 0
data_boxcox = pd.Series(boxcox(data['Customers_linear'], lmbda = 0), index = data.index)
plt.plot(data_boxcox, label="After Box Cox Transformation")
plt.legend(loc = 'best')
plt.title("Number of Customers Visisting an Ice Cream Shop since 1949")
plt.show()
```

#### Differencing
Differencing stabilises the mean of a time series by removing changes in the level of time series and therefore eliminating trend and seasonality

```python
data_boxcox_difference = pd.Series(data_boxcox - data_boxcox.shift(), index = data.index)
data_boxcox_difference.dropna(inplace = True)
plt.plot(data_boxcox_difference, label = 'After Box Cox Transformation and Differencing')

plt.legend()
plt.title('# of customers visiting an ice cream shop since 1949')
plt.show()
```

#### Checking stationarity again
#### ADF test

```python
result = adfuller(data_boxcox_difference, autolag = 'AIC')
print(f'* ADF Statistics: {result[0]}')
print(f'* p-value: {result[1]:.20f}')
print(f'* # of lags used: {result[2]:.20f}')
# print(f'* # of observations used for the ADF regression and calculation of the critical values: {result[3]}')
print('* critical values - Critical values for the test statistic at the 1%, 5%, and 10% levels. Based on MacKinnon: \n')
for key, value in result[4].items():
    print('  Critical value:', end = "")
    print(f' {key}, {value}')
```

P-value is 0 which is < 0.05 so we reject the null hypothesis (H0), and the series is now stationary.

#### Assumption 2. Autocorrelation
* Autocorrelation is the representation of the degree of similarity between a given time series and the lagged version of itself over successive time interval.  
* There are two types of Autocorrelation measures.
1. Autocorrelation function (ACF): It is a complete autocorrelation function which gives us values of auto-correlation of any series with its lagged values.
2. Partial correlation function (PACF): It gives the partial correlation of a stationary time series with its own lagged values, regressed the values of time series at all shorter lags. In contrass, Autocorrelation function doesn't control the other lags.

#### Auto Correlation function

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
# Loading and plotting ACF
plot_acf(data_boxcox_difference, ax = plt.gca(), lags = 10, title = 'Autocorrelation')
plt.xlabel('Lags')
plt.ylabel('Correlations')
plt.show()
```

#### Partial Correlation function

```python
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
# Loading and plotting PACF
plot_pacf(data_boxcox_difference, ax = plt.gca(), lags = 30, title = 'Partial Autocorrelation')
plt.xlabel('Lags')
plt.ylabel('Correlations')
plt.show()
```

## --- Applying SARIMA model to time series data ---

### Splitting into train and test of the transformed data

```python
train_data_boxcox = data_boxcox[:length_train]
test_data_boxcox = data_boxcox[length_train:]
train_data_boxcox_difference = data_boxcox_difference[:length_train - 1]
test_data_boxcox_difference = data_boxcox_difference[length_train - 1:]
```

#### Key elements of SARIMA model
* The time series is differenced to make it stationary
* The SARIMA equation is a linear combination of past observations and past errors.
* Seasonal differencing is performed on the time series.
* SARIMA model models future seasonality as a combination of linear regression of past seasonality observations and past seasonality forecast errors.

**Parameters of SARIMA model: p, q, d, m, P, Q, D**
* p - highest lag in the time series (we can determine value of p from the auto correlation function)
* q - past errors included (we can use partial autocorrelation function)
* d - degree of differencing to make the time series stationary (can be determined thru ADF and KPSS tests)
* m - the # of time steps required for a single seasonal period
* P - the seasonal autoregressive order; 0 <= P <= 4; we use grid search technique to find the optimum value
* Q - the seasonal moving average order; 0 <= Q <= 4; we use grid search technique to find the optimum value
* D - the seasonal difference order; D = 0, 1, 2; we use grid search technique to find the optimum value

### Implementing SARIMA Model

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model = SARIMAX(train_data_boxcox_difference, order=(1,1,1), seasonal_order=(1,1,1,6)) # (p,d,q) & (P,D,Q,m)
model_fit = model.fit()
print(model_fit.params)
```

### Recover the original time series data

```python
y_sarima_new = data_boxcox_difference.copy()
y_sarima_new['sarima_forecast_boxcox_difference'] = model_fit.predict(test_data_boxcox_difference.index.min(), 
                                                              test_data_boxcox_difference.index.max())
y_sarima_new['sarima_forecast_boxcox'] = y_sarima_new['sarima_forecast_boxcox_difference'].cumsum()
y_sarima_new['sarima_forecast_boxcox'] = y_sarima_new['sarima_forecast_boxcox'].add(test_data_boxcox[0])
y_sarima_new['sarima_forecast'] = np.exp(y_sarima_new['sarima_forecast_boxcox'])
```

### Plotting train, test and forecasted data

```python
plt.figure(figsize = (15, 5))

plt.plot(train['Customers_linear'], label = 'Train')
plt.plot(test['Customers_linear'], label = 'Test')
plt.plot(y_sarima_new['sarima_forecast'][test.index.min():], label = 'SARIMA model')

plt.legend()
plt.title('Seasonal Auto Regressive Integrated Moving Average Model')
plt.show()
```

### Evaluation metrics

```python
from sklearn.metrics import mean_squared_error

# Mean forecast error
mfe_SARIMA = np.round(np.mean(test['Customers'] - y_sarima_new['sarima_forecast']), 2) 

# Mean absolutely error
mae_SARIMA = np.round(np.mean(np.abs(test['Customers'] - y_sarima_new['sarima_forecast'])), 2)

# Mean absolutely percentage error
mape_SARIMA = np.round(np.mean(np.abs(test['Customers'] - y_sarima_new['sarima_forecast'])/test['Customers'])*100, 2) 

# Root mean square error
rmse_SARIMA = np.round(np.sqrt(mean_squared_error(test['Customers'], y_sarima_new['sarima_forecast'])), 2)
```

```python
result_SARIMA = pd.DataFrame({'mfe':mfe_SARIMA, 'mae':mae_SARIMA, 'mape':mape_SARIMA, 'rmse':rmse_SARIMA}, index = [0])
result_SARIMA
```
