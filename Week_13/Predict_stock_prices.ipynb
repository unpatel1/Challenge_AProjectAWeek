-- We have stock data for 506 - 1 = 505 companies

-- Apple, Amazon, IBM, Google

```python
company_list = ['AAPL_data.csv','AMZN_data.csv','IBM_data.csv','GOOG_data.csv']
```

```python
# Creating a blank dataframe to store the stock data of the above companies
all_data = pd.DataFrame()
```

```python
# Directory path of the data files
path = r'C:\Users\unpat\_Projects\Challenge_AprojectAweek_data_files\Week_13\individual_stocks_5yr'
```

```python
# Reading the data files
for file in company_list:
    current_df = pd.read_csv(filepath_or_buffer = path + '/' + file)
    
    # storing (concatenating) the data from each file into the blank dataframe
    all_data = pd.concat([all_data, current_df])
```
